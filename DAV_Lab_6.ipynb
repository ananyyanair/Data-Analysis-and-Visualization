```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('titanic.csv')
data.head()
```

```python
# (a)
# Create a bar chart to visualize the survival rate based on passenger class (Pclass).

survivalrate = data.groupby(['Pclass'])['Survived'].mean()
plt.bar(survivalrate.index, survivalrate.values, color = 'rosybrown', width = 0.5)
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.title('Survival Rate Based on Passenger Class')
plt.show()
```

```python
# (b)
# Use a pie chart to show the proportion of survivors vs. non-survivors in the dataset.

survived = data.groupby(['Survived'])['PassengerId'].count()
plt.pie(survived.values, labels = ['Not Survived', 'Survived'], colors = ['skyblue', 'palevioletred'], shadow = True, explode = [0.1, 0.1], autopct='%1.1f%%')
plt.title('Proportion of Surivors VS Non Survivors')
plt.show()
```

```python
# (c)
# Create a stacked bar chart to compare the number of survivors and non-survivors based on passenger class and sex.

group = data.groupby(['Pclass', 'Sex'])['Survived'].value_counts().unstack(fill_value = 0)
group.plot(kind = 'bar', stacked = True, color = ['salmon', 'mediumaquamarine'])
plt.xlabel('Passenger Class & Sex')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Number of Survivors')
plt.title('Number of Survivors Based on Passenger Class and Sex')
plt.show()
```
