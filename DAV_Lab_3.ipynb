```python
# (d)
# Write a function that combines the Sales Data and Customer Feedback DataFrames, 
# filling any missing Feedback_Score values in one with those from the other. 
# This is useful in cases where multiple sources have partial data for the same records
```

```python
import pandas as pd
data1 = pd.read_csv('sales data.csv')
data2 = pd.read_csv('customer feedback.csv')
def combine(data1, data2):
    combo = pd.merge(data1, data2, on='OrderID', how='outer')
    return combo
df = combine(data1, data2)
df.head()
```

```python
# (e)
# Write a function that pivots the Sales Data based on Month, with Product as rows and Sales as values, so that each monthâ€™s sales become separate
columns. This transformation can help in comparing monthly sales
trends across products.
```

```python
def pivotdata(data):
    pivoted_data = data.pivot(index='Product', columns='Month', values='Sales')
    return pivoted_data
data1 = pd.read_csv('sales data.csv')
pivotedsales = pivotdata(data1)
pivotedsales
```
