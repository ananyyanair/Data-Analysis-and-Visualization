```python
import pandas as pd
df = pd.read_csv('purchases.csv')
df
```

```python
# (a)
# Identify missing values using isnull() and notnull() functions and create a strategy for handling them with methods like fillna and dropna. 
# What impact does filling missing values in Occupation and Satisfaction_Level have on analysis results?

missing_values = df.isnull().sum()
print("Null Values - \n")
print(missing_values)

df['Occupation'].fillna(df['Occupation'].mode()[0])
df['Satisfaction Level'].fillna(df['Satisfaction Level'].mean())
df['Age'].fillna(df['Age'].mean())
df['Income'].fillna(df['Income'].mean())
df.dropna(subset=['Years Employed'])
missing_values_after = df.isnull().sum()
print("\nAfter Handling Missing Values -\n",missing_values_after)
print()
df.head()
```

```python
# (b)
# Apply custom functions to create a transformed datasetâ€”for instance, converting the Satisfaction_Level to a binary variable 
# where values above 0.7 are labeled as "High" and others as "Low."

def convert_satisfaction_level(x):
    return 'High' if x > 0.7 else 'Low'
df['Satisfaction Level'] = df['Satisfaction Level'].apply(convert_satisfaction_level)
df.head()
```

```python
# (c)
# Use map() to convert the Purchase_History categorical data into a numerical format, 
# where "High" is mapped to 2, "Medium" to 1, and "Low" to 0.

purchase_history_map = {'High': 2, 'Medium': 1, 'Low': 0}
df['Purchase History'] = df['Purchase History'].map(purchase_history_map)
df.head()
```

```python
# (d)
# Apply statistical techniques like the Z-score or IQR method to identify outliers in Income.

from scipy.stats import zscore
df['Income Z-score'] = zscore(df['Income'])
outliers = df[df['Income Z-score'].abs() > 3]
print("DataFrame with Income Z-score:")
df[['ID', 'Age', 'Gender', 'Income', 'Occupation', 'Years Employed', 'Satisfaction Level', 'Purchase History', 'Income Z-score']]
```

```python
print("\nOutliers based on Income Z-score:")
outliers[['ID', 'Age', 'Gender', 'Income', 'Occupation', 'Years Employed', 'Satisfaction Level', 'Purchase History', 'Income Z-score']]
```

```python
# (e)
# Assess the impact of missing values in Years_Employed on data reliability. 
# Apply the fillna() method using mean or median imputation.

summary_before = df['Years Employed'].describe()
df['Years Employed'].fillna(df['Years Employed'].mean())
summary_after_mean = df['Years Employed'].describe()
df = pd.read_csv('purchases.csv')  # Reset to original data
df['Years Employed'].fillna(df['Years Employed'].median())
summary_after_median = df['Years Employed'].describe()
print("Summary Before - \n", summary_before)
print("\nSummary After Mean - \n", summary_after_mean)
print("\nSummary After Median - \n", summary_after_median)
```
