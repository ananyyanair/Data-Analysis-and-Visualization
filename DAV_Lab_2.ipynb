```python
import pandas as pd
df = pd.read_csv('city.csv')
df
```

```python
# (a)
# Apply a function that fills missing values in the Age column with the
# mean age and in the City column with the value "Unknown."

df['Age'].fillna(df['Age'].mean(), inplace=True)
df['City'].fillna('Unknown', inplace=True)
df.head()
```

```python
# (b)
# Apply a function that removes duplicates based on all columns in the DataFrame.
df.drop_duplicates(inplace=True)
df
```

```python
# (c)
# Write a function that replaces inconsistent values in the Gender
# column (e.g., replace "M" with "Male" and "F" with "Female").
def replace(gender):
    if gender in ['M', 'Male']:
        return 'Male'
    elif gender in ['F', 'Female']:
        return 'Female'
    return 'Unknown'
df['Gender'] = df['Gender'].apply(replace)
df.head()
```

```python
# (d)
# d) Write a function that groups values in the Age column into ranges, such
# as "18-30," "30-40," and "40-50.
# Group Age into ranges
def age_group(age):
    if 18 <= age < 30:
        return '18-30'
    elif 30 <= age < 40:
        return '30-40'
    elif 40 <= age < 50:
        return '40-50'
    return '50+'
df['Age Group'] = df['Age'].apply(age_group)
df.head()
```
