```python
import pandas as pd
data = pd.read_csv('life expectancy.csv')
data
```

```python
# (a)
# Include functions like mean(), sum(), max(), min() from Pandas and np.mean(), np.std(), np.sum() from NumPy. 
# Demonstrate how they can be applied to quickly compute summary statistics for columns like Population and GDP in the dataset.
```

```python
import numpy as np
statspandas = {
    'Population Mean': data['Population'].mean(),
    'GDP Sum': data['GDP'].sum(),
    'GDP Max': data['GDP'].max(),
    'GDP Min': data['GDP'].min(),
}
statsnumpy = {
    'Population Mean': np.mean(data['Population']),
    'GDP Standard Deviation': np.std(data['GDP']),
    'GDP Sum': np.sum(data['GDP']),
}
stats = {
    'Pandas': statspandas,
    'NumPy': statsnumpy,
}
print("Summary Statistics:")
print("Using Pandas:")
for key, value in statspandas.items():
    print(f"  {key}: {value:,.2f}")
print("\nUsing NumPy:")
for key, value in statsnumpy.items():
    print(f"  {key}: {value:,.2f}")
```

```python
# (c)
# Use swaplevel() on the Country and Year multi-index to reverse the order of levels, then sort by Year. 
# This could be useful, for instance, when analyzing data trends over time by country rather than by year.
```

```python
multiindex = data.set_index(['Country', 'Year'])
swapped = multiindex.swaplevel().sort_index(level='Year')
swapped.head()
```
