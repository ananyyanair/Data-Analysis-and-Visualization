```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('housepricing.csv')
data.head(4)
```

```python
#  (a)
# Create a scatter plot to visualize the relationship between GrLivArea (above ground living area) and SalePrice.
plt.scatter(data['GrLivArea'], data['SalePrice'], color = 'rosybrown', marker = 'x')
plt.xlabel('Above Ground Living Area')
plt.ylabel('Sale Price')
plt.title('Relationship Between GrLivArea and SalePrice')
plt.grid()
plt.show()
```

```python
# (b)
# Use a heatmap to visualize the correlation between numerical features like GrLivArea, OverallQual, TotalBsmtSF, etc.
# plt.figure(figsize = (8, 8))
numfeatures = ['GrLivArea', 'OverallQual', 'TotalBsmtSF', 'SalePrice', 'YearBuilt']
cor = data[numfeatures].corr()
sns.heatmap(cor, cmap = 'PuBu', annot = True)
plt.title('Correlation Between Numerical Features')
plt.show()
```

```python
# (c)
# Create a bubble chart to represent the relationship between GrLivArea, SalePrice, and YearBuilt,
# where the size of the bubbles corresponds to OverallQual.
bubblesize = data['OverallQual']*10
scatter = plt.scatter(data['GrLivArea'], data['SalePrice'], c = data['YearBuilt'], s = bubblesize, cmap = 'PuBu', alpha = 0.5)
plt.colorbar(scatter, label = 'Year Built')
plt.xlabel('Above Ground Living Area')
plt.ylabel('Sale Price')
plt.title('Relationship Between GrLivArea, SalePrice and YearBuilt')
plt.grid(alpha = 0.3)
plt.show()
```
